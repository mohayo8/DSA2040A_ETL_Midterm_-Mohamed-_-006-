```python
import pandas as pd  # Import pandas for data manipulation

# Load raw data
print("Loading raw_data.csv...")
raw_df = pd.read_csv("raw_data.csv")  # Read raw_data.csv into a DataFrame
print("\nRaw Data Preview:")
print(raw_df.head())  # Display the first few rows
print("\nRaw Data Info:")
print(raw_df.info())  # Display data types and non-null counts
print("\nMissing Values:")
print(raw_df.isnull().sum())  # Check for missing values
print("\nDuplicate Rows:")
print(raw_df.duplicated().sum())  # Check for duplicate rows

# Save raw copy
raw_df.to_csv("data/raw_data.csv", index=False)  # Save the raw data to data/ folder
print("\nSaved raw_data.csv to /data directory.")

# Load incremental data
print("\nLoading incremental_data.csv...")
incremental_df = pd.read_csv("incremental_data.csv")  # Read incremental_data.csv into a DataFrame
print("\nIncremental Data Preview:")
print(incremental_df.head())  # Display the first few rows
print("\nIncremental Data Info:")
print(incremental_df.info())  # Display data types and non-null counts
print("\nMissing Values:")
print(incremental_df.isnull().sum())  # Check for missing values
print("\nDuplicate Rows:")
print(incremental_df.duplicated().sum())  # Check for duplicate rows

# Save incremental copy
incremental_df.to_csv("data/incremental_data.csv", index=False)  # Save the incremental data
print("\nSaved incremental_data.csv to /data directory.")
```
